```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
#upload data to PANDAS

fertility1920_df = pd.read_csv("fertility1920.csv")

fertility1920_df
```

```python
#Remove data from 2018 and prior
fertility1920_df = fertility1920_df.drop(labels=range(100,400))

#Remove unneeded columns
fertility1920_df = fertility1920_df.drop(columns = 'BIRTHS')
fertility1920_df = fertility1920_df.drop(columns = 'URL')

#Remove Null Data 
fertility_df.dropna(inplace=True)
fertility1920_df
```

```python
fertility1920_df.to_csv('Fertility1920_DF.csv',index=False)
```

```python
fertility22_df = pd.read_csv("fertility22.csv")

fertility22_df
```

```python
#remove unneeded data
fertility22_df = fertility22_df.drop(columns = 'birthRate')
fertility22_df = fertility22_df.drop(columns = 'totalBirths')
fertility22_df = fertility22_df.drop(columns = 'fertilityRateWhite')
fertility22_df = fertility22_df.drop(columns = 'fertilityRateBlack')
fertility22_df = fertility22_df.drop(columns = 'fertilityRateNative')
fertility22_df = fertility22_df.drop(columns = 'fertilityRateAsian')
fertility22_df
```
